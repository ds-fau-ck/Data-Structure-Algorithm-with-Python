```python
def FirstOccurrence(arr,ele,start,end):
    ind=-1
    while start<=end:
        mid=(start+end)//2
        if ele==arr[mid]:
            ind=mid
            end=mid-1
        elif ele<arr[mid]:
            end=mid-1
        else:
            start=mid+1
    return ind
def LastOccurrence(arr,ele,start,end):
    ind=-1
    while start<=end:
        mid=(start+end)//2
        if ele==arr[mid]:
            ind=mid
            start=mid+1
        elif ele<arr[mid]:
            end=mid-1
        else:
            start=mid+1
    return ind
arr=[0,1,1,5,5,5,7,7,7,7]
start=0
end=len(arr)-1
ele=7
fist_index=FirstOccurrence(arr,ele,start,end)
print(fist_index)
last_index=LastOccurrence(arr,ele,start,end)
print(last_index)
total_no=last_index-fist_index+1
print(total_no)
```
